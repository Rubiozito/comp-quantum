## Introdução ao Emaranhamento Quântico

### Definição de Emaranhamento Quântico

Emaranhamento quântico é um fenômeno peculiar da mecânica quântica onde pares ou grupos de partículas são gerados ou interagem de maneiras que o estado quântico de cada partícula não pode ser descrito independentemente dos estados das outras, mesmo que as partículas estejam separadas por grandes distâncias. Portanto, o emaranhamento implica a existência de estados quânticos globais de sistemas compostos que não podem ser escritos como um produto dos estados quânticos de subsistemas individuais. Isto
é, o conhecimento de um todo não implica conhecimento das partes.

###  Importância em Computação Quântica

* **Recursos para Computação**: O emaranhamento é um recurso chave que permite que computadores quânticos realizem cálculos que são extremamente difíceis ou impossíveis para computadores clássicos.

* **Segurança em Comunicação Quântica**: Em protocolos de criptografia quântica, como a distribuição quântica de chaves, o emaranhamento é usado para detectar eavesdroppers, garantindo a segurança das comunicações.

### Aplicações Práticas

* **Computação Quântica**: Algoritmos como o de Shor e o de Grover dependem de estados emaranhados para alcançar uma vantagem quântica sobre algoritmos clássicos.
* **Telecomunicações**: O emaranhamento é usado em experimentos de teletransporte quântico e pode potencialmente revolucionar as telecomunicações, permitindo novos tipos de protocolos de comunicação.
* **Sensores Quânticos**: Sensores que utilizam estados emaranhados podem alcançar precisões que não são possíveis com tecnologias clássicas.

###  Formalismo Matemático do Emaranhamento Quântico



O emaranhamento quântico pode ser descrito formalmente utilizando a linguagem da mecânica quântica. Um sistema de dois qubits, onde cada qubit pode estar no estado $|0\rangle$ ou $|1\rangle$, tem seu estado quântico geral representado como:

$$
|\psi\rangle = \alpha|00\rangle + \beta|01\rangle + \gamma|10\rangle + \delta|11\rangle
$$

onde$\alpha, \beta, \gamma, $ e $\delta $ são coeficientes complexos que satisfazem a condição de normalização $|\alpha|^2 + |\beta|^2 + |\gamma|^2 + |\delta|^2 = 1 $.

Um sistema físico está em um estado emaranhado se for impossível descrever esse estado composto como um produto tensorial dos estados individuais, isto é:

$$|\psi\rangle \neq\left|\psi_1\right\rangle \otimes\left|\psi_2\right\rangle \otimes \cdots \otimes\left|\psi_n\right\rangle$$

#### Exemplo de Estado Emaranhado

Consideremos um estado particular que é frequentemente utilizado para demonstrar emaranhamento:

$$
|\chi\rangle = \frac{1}{2}(|00\rangle + |01\rangle + |10\rangle - |11\rangle)
$$

Este estado é emaranhado porque não pode ser fatorado em estados de qubits individuais. A presença do termo negativo $-|11\rangle$ cria correlações quânticas entre os qubits que são essenciais para o emaranhamento.

#### Características do Estado $|\chi\rangle$

- **Não separabilidade**: O estado $|\chi\rangle$ não pode ser escrito como um produto de estados mais simples, indicando que os qubits estão emaranhados.
- **Correlações**: Se um dos qubits for medido, o resultado da medição influenciará o estado do outro qubit, uma característica fundamental dos estados emaranhados.



## Estados de Bell

### Introdução aos Estados de Bell

Os Estados de Bell são um conjunto de quatro estados quânticos específicos de dois qubits que representam exemplos clássicos de máximo emaranhamento quântico. Eles são fundamentais em muitos protocolos de informação quântica, incluindo computação, criptografia e teleportação quântica.

### Formação dos Estados de Bell

Os estados de Bell podem ser gerados a partir de um par de qubits inicialmente no estado ∣00⟩ através da aplicação de duas operações:

1. Porta Hadamard no primeiro qubit cria uma superposição.

2. Porta CNOT (Controlled-NOT): Usa o primeiro qubit como controle e o segundo como alvo, entrelaçando os dois qubits.

### Os Quatro Estados de Bell

1. $\left|\Phi^{+}\right\rangle=\frac{1}{\sqrt{2}}(|00\rangle+|11\rangle)$
2. $\left|\Phi^{-}\right\rangle=\frac{1}{\sqrt{2}}(|00\rangle-|11\rangle)$
3. $\left|\Psi^{+}\right\rangle=\frac{1}{\sqrt{2}}(|01\rangle+|10\rangle)$
4. $\left|\Psi^{-}\right\rangle=\frac{1}{\sqrt{2}}(|01\rangle-|10\rangle)$

### Exemplo Prático no Qiskit: Preparação dos Estados de Bell 

**Passo 1**: Importar Bibliotecas Necessárias

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere
from qiskit.quantum_info import Statevector
#import matplotlib.pyplot as plt
```

**Passo 2**: Preparação e Visualização de um Estado de Bell $\left|\Phi^{+}\right\rangle$

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere

# Cria um circuito quântico com 2 qubits
circuite = QuantumCircuit(2,2)

# Aplica uma porta Hadamard no primeiro qubit
circuite.h(0)

# Aplica uma porta CNOT com o primeiro qubit como controle e o segundo como alvo
circuite.cx(0, 1)

print(circuite.draw())

# Preparar o estado quântico
state = Statevector.from_instruction(circuite)

# Visualizar o estado no Q-sphere
plot_state_qsphere(state)
```

No Q-sphere, cada ponto representa um termo na superposição do estado quântico, com a posição do ponto indicando as fases relativas entre os estados. Vejamos cada componente da imagem:

- **Esfera**: A esfera serve como uma superfície de representação onde os estados quânticos de dois qubits são projetados.
- **Pontos Azuis**: Cada ponto representa um componente do estado quântico. Aqui, você vê dois pontos, um no topo e um na parte inferior da esfera.
- O ponto superior representa o estado $|00\rangle$.
- O ponto inferior representa o estado $|11\rangle$.
- **Linha Conectando os Pontos**: A linha e a sua cor, entre os dois pontos, indica que esses dois componentes estão em fase (ou seja, não há diferença de fase angular entre eles).

 ### Influência das Medições no Emaranhamento

**Objetivo**: Explorar o colapso de estados entrelaçados mediante medições e entender a não-localidade quântica.

Para entender o comportamento, vamos analisar dois casos: um em que o estado não está entrelaçado e outro que apresenta entrelaçamento.

#### Estado não-emaranhado

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere

# Cria um circuito quântico com 2 qubits
circuit_ne = QuantumCircuit(2,2)

# Aplica uma porta Hadamard no primeiro qubit
circuit_ne.h(0)
circuit_ne.h(1)

# Aplica uma porta CNOT com o primeiro qubit como controle e o segundo como alvo
#circuit.cx(0, 1)

print(circuit_ne.draw())

# Preparar o estado quântico
state = Statevector.from_instruction(circuit_ne)

# Visualizar o estado no Q-sphere
plot_state_qsphere(state)
```

**Note que os estados não estão emaranhados**:

$$
|\psi\rangle = \alpha|00\rangle + \beta|01\rangle + \gamma|10\rangle + \delta|11\rangle
$$

Executando o circuito e medindo os dois qubits (q$_0$ e q$_1$) observa-se o seguinte resultado:

```python
# Adiciona medições ao circuito
circuit_ne.measure([0, 1], [0, 1])
print(circuit_ne.draw())

# Escolher o simulador
simulator = Aer.get_backend('qasm_simulator')

# Transpilar o circuito para o simulador
transpiled_circuit_ne = transpile(circuit_ne, simulator)

# Executar o circuito transpilado
job = simulator.run(transpiled_circuit_ne, shots=1000)

# Obter os resultados
result = job.result()

# Obter as contagens
counts = result.get_counts()

# Garantir que todas as combinações possíveis estejam no histograma
all_possible_outcomes = ['00', '01', '10', '11']
for outcome in all_possible_outcomes:
    if outcome not in counts:
        counts[outcome] = 0

# Visualizar os resultados
#plot_histogram(counts)

# Desagregar contagens para cada qubit individualmente
qubit0_counts = {'0': 0, '1': 0}
qubit1_counts = {'0': 0, '1': 0}

for outcome in counts:
    qubit0_counts[outcome[0]] += counts[outcome]
    qubit1_counts[outcome[1]] += counts[outcome]

# Plotar histogramas separados para cada qubit
plot_histogram([qubit0_counts, qubit1_counts], legend=['Qubit 0', 'Qubit 1'])
```

### Conclusões e Implicações 

Se os qubits estão emaranhados de uma forma específica, como no estado de Bell $\left|\Phi^{+}\right\rangle=$ $\frac{1}{\sqrt{2}}(|00\rangle+|11\rangle)$, então a medição de um qubit influencia diretamente o estado do outro devido à natureza do emaranhamento. No caso do estado $\left|\Phi^{+}\right\rangle$, se você medir o qubit 0 e encontrar que ele está no estado $|0\rangle$, então o qubit 1 também será encontrado no estado $|0\rangle$ devido ao emaranhamento entre eles.

1. **Não Localidade**: O fenômeno do emaranhamento é um exemplo da "ação fantasmagórica à distância" descrita por Einstein, onde a medição de um qubit instantaneamente afeta o estado do outro, independentemente da distância entre eles.

2. **Uso em Computação Quântica e Informação Quântica**: O emaranhamento é usado em protocolos avançados como a teleportação quântica, computação quântica distribuída e criptografia quântica.
